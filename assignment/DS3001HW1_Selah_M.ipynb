```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x= np.linspace(0,1,num=50)

log_array = np.log(x)
print(log_array)

exponent_array = np.exp(x)
print(exponent_array)


```

```python
plt.scatter(x, log_array, label='Natural Log')
plt.scatter(x, exponent_array, label='Exponential')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc = 'lower right')
plt.title("Natural Log and Exponential Functions")

plt.show()
```

```python
x = np.linspace(-6.5,6.5,num=100)
sin_array = np.sin(x)
cos_array = np.cos(x)
```

```python
plt.plot(x, sin_array, label='Sine')  
plt.plot(x, cos_array, label='Cosine') 
plt.xlabel("x values")
plt.ylabel("function values")
plt.legend(loc = 'lower left')
plt.title("Cosine and Sine")

plt.show()
```
